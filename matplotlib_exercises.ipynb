```python
plt.figure(figsize = (8,2))

x1 = np.arange(50)
square_root = x**0.5

# plt.xlim(0, max(x))
# plt.ylim(0, max(y))
plt.plot(x1,square_root)
plt.show()
```

```python
x1 = np.array(range(10))
y1 = x1**.5

plt.plot(x1,y1)
plt.title('$y = \sqrt{x}$')
plt.show()
```

```python
plt.figure(figsize = (8,2))

x2 = range(-5,5)
cube = x**3
# y = func(x)

# plt.xlim(0, max(x))
# plt.ylim(0, max(y))
plt.plot(x2,cube)
plt.show()
```

```python
plt.figure(figsize = (8,2))

x3 = np.arange(50)
exponent = x * 2

# plt.xlim(0, max(x))
# plt.ylim(0, max(y))
plt.plot(x3,exponent)
plt.show()
```

```python
x2 = np.array(range(-10,10))
y2 = x2**3

plt.plot(x2,y2)
plt.title('$y = x^3$')
plt.show()
```

```python
x3 = np.array(range(10))
y3 = 2**x3

plt.plot(x3,y3)
plt.title('$y = 2^x$')
plt.show()
```

```python
plt.figure(figsize = (8,2))

x4 = np.arange(50)
denominator = 1 // (x + 1)

# plt.xlim(0, max(x))
# plt.ylim(0, max(y))
plt.plot(x4,denominator)
plt.show()
```

```python
x4 = np.linspace(-10,10,100)
# x4 = np.arange(-10,10)
y4 = 1 / (1+x4)

plt.plot(x4,y4)
plt.title('$y = 1/(x + 1)$')
plt.show()
```

```python
# Combine the figures you created in the last step into one large figure with 4 subplots.
```

```python
plt.figure(figsize = (8,2))

plt.subplot(2, 2, 1)
plt.text(0.5, 0.5, str(221))
plt.plot(x1, square_root)

plt.subplot(222)
plt.text(0.5, 0.5, str(222))
plt.plot(x2, cube)

plt.subplot(223)
plt.text(0.5, 0.5, str(223))
plt.plot(x3, exponent)

plt.subplot(224)
plt.text(0.5, 0.5, str(224))
plt.plot(x4, denominator)
```

```python
fig, ax = plt.subplots(2, 2, figsize=(12, 6))

plt.subplot(221)
plt.plot(x1,y1)
plt.title('$y = \sqrt{x}$')


plt.subplot(222)
plt.plot(x2,y2)
plt.title('$y = x^3$')


plt.subplot(223)
plt.plot(x3,y3)
plt.title('$y = 2^x$')


plt.subplot(224)
plt.title('$y = 1/(x + 1)$')
plt.plot(x4,y4)

fig.suptitle('my four plots separate')
plt.tight_layout()
plt.show()
```

```python
# Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. 
# Be sure to include a legend and an appropriate title for the figure.
```

```python
#create figure size
plt.figure(figsize=(8,6))

#plot square root
plt.plot(x, square_root, color='gray', ls = '--', marker = '^', label='Square Root')

#plot cube
plt.plot(x, cube, color = '#33FF8E', lw = 2, label = 'Cube')

#plot exponent
plt.plot(x, exponent, color = 'red', lw = 2, label = 'Exponent')

#plot denominator
plt.plot(x, denominator, color = 'yellow', lw = 3, label = 'Denominator')


#add legend with location
plt.legend(loc = 'upper center')

#add title
plt.title('Y for X Equations')

#add x and y labels
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

#add gridlines
plt.grid(ls='-.')

#add annotations
# plt.annotate('This Salary', xy=(25, 38000), xytext=(32,40000), arrowprops=dict(facecolor='gray'))

#finalize
plt.show()
```
